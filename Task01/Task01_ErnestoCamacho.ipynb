**2) Calcular y mostrar la cantidad de pacientes recuperados y muertos en cada continente.**
- Para este ejercicio los paises se van a organizar de acuerdo a un csv encontrado en internet. (https://github.com/dbouquin/IS_608/blob/master/NanosatDB_munging/Countries-Continents.csv)
- Esto significa que los datos que no son de paises son eliminados

```python
continent_dataframe= pd.read_csv("https://raw.githubusercontent.com/ErnestoCamachoA9805/MIND/main/Task01/Continents.csv")
dataf= covid_gobal_dataframe.merge(continent_dataframe, on="Country")
dataf= dataf.sort_values(["Continent","Country"])
continent_dataframe= None
## 
dir= {"Africa":[0,0],"Asia":[0,0],"Europe":[0,0],"North America":[0,0],"Oceania":[0,0],"South America":[0,0],}
for index, row in dataf.iterrows():
  if row["status"]== "D":
    dir[row["Continent"]][0]+= row["12/31/21"] + row["12/31/20"]
  elif row["status"]== "R":
    dir[row["Continent"]][1]+= row["12/31/21"] + row["12/31/20"]
for key in dir.keys():
  print("El numero de muertos en el contiente " + key + " es: " + str(dir[key][0]))
  print("El numero de recuperaciones en el contiente " + key + " es: " + str(dir[key][1]))
  print()
```

- Los resultados salen de la suma de 2 fechas que considero reportan de manera completa la cantidad de personas por año, esta fecha es el 31 de diciembre del año 

**3) Indicar el país con el mayor crecimiento en la cantidad de recuperados y muertos durante el año 2022.**

- Un detalle de los datos es que no hay datos de recuperaciones en el 2022 por lo que se hara el calculo solo las muertes. 
- Los datos de poblacion de cada pais son recuperados del csv de la pagina(https://worldpopulationreview.com/)

```python
pop_data= pd.read_csv("https://raw.githubusercontent.com/ErnestoCamachoA9805/MIND/main/Task01/world_population.csv")
pop_data= pop_data[['name','pop2021']]
pop_data= pop_data.rename(columns={'name': 'Country'})
dataf= dataf.merge(pop_data, on="Country")
```

```python
temporalDf= dataf[dataf["status"]!= "R"]
country= ''
value= 0
for index, row in temporalDf.iterrows():
  tasa= (row["01/18/22"] - row["01/01/22"])/row["pop2021"]
  if tasa >= value:
    country= row["Country"]
print("El pais con mayor crecimiento en muertos del 2022 es: " + country)
```

- Para determinar el valor tome el valor de la fecha mas reciente, menos el valor del primero de enero, dividio la poblacion del pais.

**4) Indicar el país con el menor crecimiento en la cantidad de recuperados y muertos durante el año 2021.**

```python
countryR= ''
countryD= ''
valueR= Inf
valueD= Inf
for index, row in dataf.iterrows():
  tasa= (row["12/31/21"] - row["01/01/21"])/row["pop2021"]
  if row["status"] == "R":
    if tasa <= valueR:
      #print(row["Country"])
      countryR= row["Country"]
  else:
    if tasa <= valueD:
      countryD= row["Country"]
print("El pais con menor crecimiento en recuperados del 2021 es: " + countryR)
print("El pais con menor crecimiento en muertos del 2021 es: " + countryD)    
```
